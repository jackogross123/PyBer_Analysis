```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Set the x-axis equal to a list of months within a year
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]
# Set the y-axis equal to a list of floats as the total fare in US dollars acccumulated each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
plt.xlabel("Months")
plt.ylabel("Revenue $")
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the limit between 0 and 45.
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color='blue', linewidth=2, label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the limit between 0 and 45.
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add the grid
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="+", color='green', linewidth=2, label='Boston')
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title
ax.set_title("PyBer Fare by Month")
# Add the grid
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Create the bar
plt.bar(x_axis, y_axis)
```

```python
# Create the bar graph.
plt.bar(x_axis, y_axis, color='green', label='Boston')
# Create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create the title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
# Invert the yaxis to Jan is at the top.
plt.gca().invert_yaxis()
# Add title, xaxis, and yaxis labels
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax=plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
# Invert yaxis
ax.invert_yaxis()
# Add title and axes labels
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_title("PyBer Fare by Month")
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(y_axis, x_axis, c='red', label="Chicago")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s =[i * 3 for i in y_axis])
```

```python
# Create a scatter plot using the object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble plot using the object-oriented interface
fig, ax=plt.subplots()
ax.scatter(y_axis, x_axis, s =[i * 3 for i in y_axis], label='Boston', c='skyblue', edgecolors='black', linewidth=2, alpha=0.80)
ax.invert_yaxis()
# Add titles and labels
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_title("PyBer Fare by Month")
# Set xlimit
ax.set_xlim(0, 50)
# Add legend.
ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a pie chart using Object oriented
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    shadow=True,
    counterclock=False,
    startangle=90)

plt.show()
```
